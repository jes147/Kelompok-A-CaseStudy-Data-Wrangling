# Interpretasi 04

Di sini, kami menggunakan nilai rata rata pada tiap mahasiswa yang ada. Pada data terdapat perbedaan kemampuan akademik antara laki-laki terhadap perempuan. Rata-rata nilai pada gender laki-laki (pria) = 81.2222 dan perempuan (wanita) = 74.83333. Hal ini berarti kemampuan akademik laki-laki lebih baik dibandingkan perempuan.

```python
#Code soal No 5

#merge dataframe nilai dan mahasiswa dari kolom nim
nilai_mahasiswa=pd.merge(nilai,mahasiswa, on='nim')
nilai_mahasiswa=nilai_mahasiswa[['Angkatan', 'id_mata_kuliah','nilai']]

#mencari mean
mean =nilai_mahasiswa.groupby(['Angkatan','id_mata_kuliah']).mean()[['nilai']]
mean =mean.rename({'nilai':'mean_nilai'}, axis=1)

#mencari standar deviasi
sb = nilai_mahasiswa.groupby(['Angkatan','id_mata_kuliah']).std()[['nilai']]
sb = sb.rename({'nilai':'std_nilai'}, axis=1)

#mencari min
min = nilai_mahasiswa.groupby(['Angkatan','id_mata_kuliah']).min()[['nilai']]
min = min.rename({'nilai':'min_nilai'}, axis=1)

#mencari max
max = nilai_mahasiswa.groupby(['Angkatan','id_mata_kuliah']).max()[['nilai']]
max = max.rename({'nilai':'max_nilai'}, axis=1)

nilai_mahasiswa = nilai_mahasiswa.set_index(['Angkatan'])
nilai_mahasiswa = pd.concat([mean, sb, min, max], axis=1)
nilai_mahasiswa
```
